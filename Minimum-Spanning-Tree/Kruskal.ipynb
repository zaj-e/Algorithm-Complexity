```python
class Disjointset:
    def __init__(self, n):
        self.N = n
        self.sets = [i for i in range(n)]
        
    def Find(self, a):
        x = a
        while x != self.sets[x]:
            x = self.sets[x]
        return x
    
    def Union(self, a, b):
        pa = self.Find(a)
        pb = self.Find(b)
        if pa != pb:
            self.sets[pa] = pb
            
    def IsSameSet(self, a, b):
        return self.Find(a) == self.Find(b)
```

```python
def Kruskal(G):
    n =len(G)
    queue =[]
    for u in range(n):
        for v, w in G[u]:
            hq.heappush(queue, (w, u, v)) 
            
    ds = Disjointset(n)
    Tree =[-1]*n
    while n>1:
        _,u,v = hq.heappop(queue)
        if not ds.IsSameSet(u,v):
            ds.Union(u,v)
            Tree[v]=u
            n -=1
    return Tree    
```

```python
G = [[(1, 1), (2, 5)],
     [(0, 1), (4, 6), (5, 7)],
     [(0, 5), (3, 3), (4, 2)],
     [(2, 3), (4, 4), (5, 1)],
     [(1, 6), (2, 2), (3, 4), (5, 3)],
     [(1, 7), (3, 1), (4, 3)]]
```

```python
def Dot(T):
    n = len(T)
    dot = Graph()
    for i in range(n):
        dot.node(str(i))
    for i in range(n):
        if T[i] >= 0:
            dot.edge(str(i), str(T[i]))
    dot.graph_attr['rankdir'] = 'BT'
    return dot
```

```python
T = Kruskal(G)
print(T)
```
