```python
from graphviz import Digraph, Source
```

```python
class Dottable:
    def Dot(self):
        dot = Digraph()
        for i in range(self.N):
            dot.node(str(i))
        for i in range(self.N): #1:20 por ahi
            if self.sets[i]!=i:
                dot.edge(str(i),str(self.sets[i]))
        dot.graph_attr['rankdir']='BT'
        return dot
    pass
```

```python
class DisjoinSetQF(Dottable):
    def __init__(self,n):
        self.N = n
        self.sets = [i for i in range(n)]
    def Find(self, a):
        return self.sets[a]
    def Union(self,a,b):
        pa = self.Find(a)
        pb = self.Find(b)
        for i in range(self.N):
            if self.sets[i] == pa:
                self.sets[i] = pb
```

```python
ds = DisjoinSetQF(10)
```
