```python
class Dottable:
    def Dot(self):
        dot = Digraph()
        for i in range(self.N):
            dot.node(str(i))
        for i in range(self.N):
            if self.sets[i] != i:
                dot.edge(str(i),str(self.sets[i]))
        dot.graph_attr['rankdir'] = 'BT'
        return dot
```
