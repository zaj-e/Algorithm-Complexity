```python
from graphviz import Digraph, Source
```

```python
class Dottable:
    def Dot(self):
        dot = Digraph()
        for i in range(self.N):
            dot.node(str(i))
        for i in range(self.N):
            if self.sets[i] != i:
                dot.edge(str(i),str(self.sets[i]))
        dot.graph_attr['rankdir'] = 'BT'
        return dot
```

```python
class DisjoinSetPc(Dottable):
    def __init__(self, n):
        self.N = n
        self.sets = [i for i in range(n)]
    def Find(self, x):
        if x == self.sets[x]:
            return x
        else:
            antepasado = self.Find(self.sets[x])
            self.sets[x] = antepasado
            return antepasado
    def Union(self, a, b):
        pa = self.Find(a)
        pb = self.Find(b)
        if pa != pb:
            self.sets[pb] = pa
```

```python
ds = DisjoinSetPc(10)
ds.Dot()
```
